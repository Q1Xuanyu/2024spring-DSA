```python
from collections import deque
t=int(input())
queue=deque()
dic={}
for i in range(t):
    for j in input().split():
        dic[j]=i
dickeys=dic.keys()

while True:
    com=input()
    if com=='STOP':
        break
    elif com=='DEQUEUE':
        print(queue.popleft())
    else:
        mem=com.split()[1]
        if mem in dickeys:
            check=1
            Queue=list(queue)
            for i in range(len(Queue)-1,-1,-1):
                if dic[Queue[i]]==dic[mem]:
                    queue.insert(i+1,mem)
                    check=0
                    break
            if check:
                queue.append(mem)
        else:
            queue.append(mem)
```

```python
from collections import deque					# 时间: 105ms

# Initialize groups and mapping of members to their groups
t = int(input())
groups = {}
member_to_group = {}



for _ in range(t):
    members = list(map(int, input().split()))
    group_id = members[0]  # Assuming the first member's ID represents the group ID
    groups[group_id] = deque()
    for member in members:
        member_to_group[member] = group_id

# Initialize the main queue to keep track of the group order
queue = deque()
# A set to quickly check if a group is already in the queue
queue_set = set()


while True:
    command = input().split()
    if command[0] == 'STOP':
        break
    elif command[0] == 'ENQUEUE':
        x = int(command[1])
        group = member_to_group.get(x, None)
        # Create a new group if it's a new member not in the initial list
        if group is None:
            group = x
            groups[group] = deque([x])
            member_to_group[x] = group
        else:
            groups[group].append(x)
        if group not in queue_set:
            queue.append(group)
            queue_set.add(group)
    elif command[0] == 'DEQUEUE':
        if queue:
            group = queue[0]
            x = groups[group].popleft()
            print(x)
            if not groups[group]:  # If the group's queue is empty, remove it from the main queue
                queue.popleft()
                queue_set.remove(group)
```

```python
from collections import deque
import sys
input = sys.stdin.read

def simulate_queue():
    data = input().strip().split('\n')
    idx = 0
    
    t = int(data[idx])
    idx += 1
    
    group_map = {}
    
    # 记录每个人属于哪个小组
    for i in range(t):
        for member in data[idx].split():
            group_map[int(member)] = i
        idx += 1
    
    # 队列模拟
    queue = deque()  # 整体队列，用于记录小组的顺序
    group_queues = {}  # 各个小组的队列，记录小组内的成员顺序
    results = []
    
    while idx < len(data):
        command = data[idx]
        idx += 1
        
        if command == 'STOP':
            break
        
        if command.startswith('ENQUEUE'):
            person = int(command.split()[1])
            group_id = group_map.get(person, None)  # 获取小组id，可能是None
            
            # 检查此人的小组是否已在队列中
            if group_id is not None:
                if group_id not in group_queues:
                    group_queues[group_id] = deque()
                    queue.append(group_id)
                group_queues[group_id].append(person)
            else:
                # 独立的人，没有小组，当作单独的小组处理
                if person not in group_queues:
                    group_queues[person] = deque([person])
                    queue.append(person)
                
        elif command == 'DEQUEUE':
            if queue:
                first_group = queue[0]
                results.append(group_queues[first_group].popleft())
                if not group_queues[first_group]:
                    queue.popleft()
                    del group_queues[first_group]
    
    return results

# 输出结果
for result in simulate_queue():
    print(result)
```
